```python
# --- Support Functions ---
# This section contains support functions used by the main code.

def read_input(file_path):
    with open(file_path, 'r') as file:
        return file.read().splitlines()
    
```

```python
# --- Part ONE ---

input = read_input('input.txt')

result = 0

calc_list = input[0].split()
operations = input[-1].split()
#print (calc_list)

for line in input[1:-1]:
    line = line.split()
    calc_list = [str(eval(calc_list[i] + operations[i] + line[i])) for i in range(len(calc_list))]
    #print (calc_list)

calc_list = [int(x) for x in calc_list]

result = sum(calc_list)

print("Part ONE:", result)
```

```python
# --- Part TWO ---
input = read_input('input.txt')

result = 0
# calc the  width of each column


nr_digits = len(input)-1
width = len(input[0])
operators = input[-1]

calc_list = []

calc = []
for i in range(width):
    number = 0
    if operators[i] != ' ':
        calc = [operators[i]]
    digits = [input[j][i] for j in range(nr_digits) if input[j][i] != ' ']
    if digits:
        calc.append(''.join(digits))
    else:
        to_eval = calc[0].join(calc[1:])
        calc_list.append(eval(to_eval))
        calc = []

to_eval = calc[0].join(calc[1:])
calc_list.append(eval(to_eval))

result = sum(calc_list)

print("Part TWO:", result)
```
