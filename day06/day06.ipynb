```python
# --- Support Functions ---
# This section contains support functions used by the main code.

def read_input(file_path):
    with open(file_path, 'r') as file:
        return file.read().splitlines()
    
```

```python
# Function to sum two lists element-wise
def sum_lists(list1, list2):
    if len(list1) != len(list2):
        raise ValueError("Lists must have the same length")
    return [a + b for a, b in zip(list1, list2)]
```

```python
# --- Part ONE ---

input = read_input('input.txt')

result = 0

calc_list = input[0].split()
operations = input[-1].split()
#print (calc_list)

for line in input[1:-1]:
    line = line.split()
    calc_list = [str(eval(calc_list[i] + operations[i] + line[i])) for i in range(len(calc_list))]
    #print (calc_list)

calc_list = [int(x) for x in calc_list]

result = sum(calc_list)

print("Part ONE:", result)
```
