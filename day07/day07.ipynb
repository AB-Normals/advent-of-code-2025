```python
# --- Support Functions ---
# This section contains support functions used by the main code.

def read_input(file_path):
    with open(file_path, 'r') as file:
        return file.read().splitlines()
    
```

```python
# --- Part ONE ---

input = read_input('input.txt')

result = 0

# print (input[0])
for idx in range(1,len(input)):
    l = [c for c in input[idx]]
    for i in range(len(l)):
        c0 = input[idx-1][i]
        c = l[i]
        if c0 == 'S' or c0 == '|':  # a Beam or the Source in previous line
            if c == '^': # splitter
                l[i-1] = '|'
                l[i+1] = '|'
                result += 1
            else:
                l[i] = '|'
    input[idx] = "".join(l)
    # print (input[idx])

print("Part ONE:", result)
```

```python
# --- Part TWO ---
input = read_input('input.txt')

result = 0

cache = {}

def path_count(row, col, input=input):
    if row == len(input)-1:
        return 1
    if (row, col) in cache:
        return cache[(row, col)]

    retvalue = 0
    if input[row][col] == '^':
        retvalue = (path_count(row + 1, col - 1, input) + path_count(row + 1, col + 1, input))
    else:
        retvalue = path_count(row + 1, col, input)
    cache[(row, col)] = retvalue
    return retvalue

# For this resolution, follow all possible path of the beams starting from source S
# path is a list of tuples (row, col)
start = (0, input[0].index('S'))  # (row, col)

result = path_count(start[0], start[1], input)


print("Part TWO:", result)
```
